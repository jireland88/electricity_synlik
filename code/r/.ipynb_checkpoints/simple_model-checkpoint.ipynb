# Simple Model

```r
library(nlme)
library(mvnfast)

source("models.r")
source("helper_functions.r")


stats <- function(xs) {
    sat <- extract_days(xs, 5)
    sun <- extract_days(xs, 6)
    
    reg <- autoregress(xs, 1)
    
    return(c(mean(xs), sd(xs), mean(sat), mean(sun), reg[1], reg[2]))
}


loss <- function(s, xs) {
    sat <- extract_days(xs, 5)
    sun <- extract_days(xs, 6)
    
    return(-sum(dnorm(xs, s[1], s[2], log=TRUE)) +
           0.5 * sum((sat - s[3])^2) +
           0.5 * sum((sun - s[4])^2) +
           loss_autoregress(xs, 1, c(s[5], s[6])))
}
```

```r
obs_length <- 100
true_par <- c(2.852, -0.089, -0.192, log(0.112 / (1 - 0.112)), log(0.114))

obs <- do.call("rjm_simple", as.list(append(true_par, obs_length)))
s_obs <- stats(obs)

loss(s_obs, obs)
```

```r
calc_hessian <- function(s, xs) {
    H <- matrix(0, 6, 6)
    
    m <- length(xs)
    
    H[1, 1] <- m / (s[2]^2)
    H[1, 2] <- 2 * sum(xs - s[1]) / (s[2]^3)
    H[2, 1] <- 2 * sum(xs - s[1]) / (s[2]^3)
    H[2, 2] <- -(m/(s[2]^2)) + (3 * sum((xs - s[1])^2) / (s[2]^4))
    
    H[3, 3] <- floor((m + 2) / 7)
    H[4, 4] <- floor((m + 1) / 7)
    
    H[5,5] <- 2 * (m - 1)
    H[5,6] <- 2 * sum(xs[2:m])
    H[6,5] <- 2 * sum(xs[2:m])
    H[6,6] <- 2 * sum(xs[2:m]^2)
    
    return(H)
}

calc_gradient <- function(s, xs) {
    n = length(xs)
    idx = 1:n
    
    gr <- cbind(-(xs - s[1]) / s[2]^2,
                1/s[2] - ((xs - s[1])^2 / s[2]^3),
                (s[3] - xs)*((idx-1) %% 7 == 5),
                (s[4] - xs)*((idx-1) %% 7 == 6),
                c(0, (-2*xs[2:n]) + (2*s[6]*xs[1:(n-1)]) + 2*s[5]),
                c(0, xs[1:(n-1)]*((-2*xs[2:n]) + (2*s[6]*xs[1:(n-1)]) + 2*s[5])))
    
    return(cov(gr)*n)
}
```

```r
synthetic_likelihood <- function(pars, obs_length, s_obs) {
    
    sim <- do.call("rjm_simple", as.list(append(pars, obs_length)))
  
    s_sim <- stats(sim)
    
    H <- calc_hessian(s_sim, sim)
    V <- calc_gradient(s_sim, sim)
    
    Sigma <- solve(H %*% solve(V, H))
    
    out <- dmvn(s_obs, s_sim, Sigma, log = TRUE)
  
  return(out)
}

synthetic_likelihood(true_par, obs_length, s_obs)
```

```r
library(synlik)
```

```r
likfun <- function(params) {
  return(synthetic_likelihood(params, obs_length, s_obs))
}

print(likfun(true_par))

res <- synlik:::ml(likFun = likfun,
                   initPar = true_par,
                   initCov = diag(c(1, 1, 1, 1, 1))^2,
                   np = 50,
                   niter = 200)
```

```r
options(repr.plot.width = 10, repr.plot.height = 10)
par(mfrow = c(3, 2))
for (i in 1:5) {
    plot(res$estim[, i])
    abline(h = true_par[i], col = 2)
}
```

```r
# Simulation Study
M = 50
final_params <- matrix(0, nrow=M, ncol=length(true_par))
for (j in 1:M) {
  print(j)
  
  obs <- do.call("rjm_simple", as.list(append(true_par, obs_length)))
  s_obs <- stats(obs)
  
  likfun <- function(params) {
    return(synthetic_likelihood(params, obs_length, s_obs))
  }
  
  res <- synlik:::ml(likFun = likfun,
                     initPar = true_par,
                     initCov = diag(c(1, 1, 1, 1, 1))^2,
                     np = 50,
                     niter = 200)
  
  
  final_params[j, ] <- res$estim[200, ]
}
```

```r
synthetic_likelihood <- function(pars, N, obs_length, s_obs) {
  
  stats_matrix <- matrix(0, nrow=N, ncol=length(s_obs))
  
  for (i in 1:N) {
    sim <- do.call("rjm_simple", as.list(append(pars, obs_length)))
    stats_matrix[i, ] = stats(sim)
  }
  
  mu <- colMeans(stats_matrix)
  Sigma <- cov(stats_matrix)
  
  out <- dmvn(s_obs, mu, Sigma, log = TRUE)
  
  return(out)
}

N <- 100
synthetic_likelihood(true_par, N, obs_length, s_obs)
```

```r
library(microbenchmark)
library(ggplot2)
library(cowplot)

synthetic_likelihood <- function(pars, obs_length, s_obs) {
    
    sim <- do.call("rjm_simple", as.list(append(pars, obs_length)))
  
    s_sim <- stats(sim)
    
    H <- calc_hessian(s_sim, sim)
    V <- calc_gradient(s_sim, sim)
    
    Sigma <- solve(H %*% solve(V, H))
    
    out <- dmvn(s_obs, s_sim, Sigma, log = TRUE)
  
  return(out)
}

synthetic_likelihood_no_robust <- function(pars, N, obs_length, s_obs) {
  
  stats_matrix <- matrix(0, nrow=N, ncol=length(s_obs))
  
  for (i in 1:N) {
    sim <- do.call("rjm_simple", as.list(append(pars, obs_length)))
    stats_matrix[i, ] = stats(sim)
  }
  
  mu <- colMeans(stats_matrix)
  Sigma <- cov(stats_matrix)
  
  out <- dmvn(s_obs, mu, Sigma, log = TRUE)
  
  return(out)
}

library(nlme)

synthetic_likelihood_numeric <- function(pars, obs_length, s_obs) {
    
    sim <- do.call("rjm_simple", as.list(append(pars, obs_length)))
  
    s_sim <- stats(sim)
    
    H <- fdHess(s_sim, loss, xs = sim)$Hessian
    V <- calc_gradient(s_sim, sim)
    
    Sigma <- solve(H %*% solve(V, H))
    
    out <- dmvn(s_obs, s_sim, Sigma, log = TRUE)
  
  return(out)
}

N <- 100

no_robust_test <- function() {
    obs <- do.call("rjm_simple", as.list(append(true_par, obs_length)))
    s_obs <- stats(obs)
    
    synthetic_likelihood_no_robust(true_par, N, obs_length, s_obs)
}

robust_test <- function() {
    obs <- do.call("rjm_simple", as.list(append(true_par, obs_length)))
    s_obs <- stats(obs)
    
    synthetic_likelihood(true_par, obs_length, s_obs)
}

robust_numeric_test <- function() {
    obs <- do.call("rjm_simple", as.list(append(true_par, obs_length)))
    s_obs <- stats(obs)
    
    synthetic_likelihood_numeric(true_par, obs_length, s_obs)
}



bm <- microbenchmark(" " = no_robust_test())

p1 <- autoplot(bm) +
    ylab("") +
    labs(title = "Performance (No RCM)") +
    theme_half_open() +
    theme(plot.title = element_text(hjust = 0.5))

m_no_robust <- median(bm$time)

bm <- microbenchmark(" " = robust_test())

p2 <- autoplot(bm) +
    labs(title = "Performance (With Analytic RCM)") +
    theme_half_open() +
    theme(plot.title = element_text(hjust = 0.5))

m_robust <- median(bm$time)

bm <- microbenchmark(" " = robust_numeric_test())

p3 <- autoplot(bm) +
    labs(title = "Performance (With Numerical RCM)") +
    theme_half_open() +
    theme(plot.title = element_text(hjust = 0.5))

m_robust_num <- median(bm$time)

options(repr.plot.width=20, repr.plot.height=8)
plot_grid(p1, p3, p2, nrow=1, ncol=3)

print(m_no_robust / m_robust)
print(m_no_robust / m_robust_num)
```
